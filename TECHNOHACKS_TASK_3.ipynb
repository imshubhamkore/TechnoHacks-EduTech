## Exploratory Data Analysis

```python
# Create a scatter plot with 'SepalLengthCm' on the x-axis and 'SepalWidthCm' on the y-axis
# Differentiate data points by 'Species' using different colors
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm",data = data,hue = 'Species')
plt.show()
```

```python
# Create a scatter plot with 'PetalLengthCm' on the x-axis and 'PetalWidthCm' on the y-axis
# Differentiate data points by 'Species' using different colors
sns.scatterplot(x="PetalLengthCm", y="PetalWidthCm",data = data ,hue = 'Species')
plt.show()
```

```python
# Loop through each column in the DataFrame except the last one
for col in data.columns[ :-1]:
    # Create a box plot with 'Species' on the x-axis and the current column 'col' on the y-axis
    sns.boxplot(x="Species", y=col, data=data)
    # Display the plot
    plt.show()
```

```python
# Loop through each column in the DataFrame except the last one
for col in data.columns[ :-1]:
    # Create a histogram plot with the current column 'col' on the x-axis
    sns.histplot(x=col, data=data)

    # Display the plot
    plt.show()
```

```python
# Setosa is the most separable.
# Create a pair plot for the entire dataset, with data points colored by 'Species' using a Dark2 color palette
sns.pairplot(data,hue='Species',palette='Dark2')
```
